## DataFrame y Series

Un DataFrame es una **tabla compuesta por arreglos de entradas individuales**, cada entrada con un cierto valor.

A cada entrada le corresponde una **fila y una columna**. \
El constructor *pd.DataFrame* recibe un **diccionario de listas** como argumento. \
Las llaves del diccionario se asignan como los nombres de las columnas y las listas se asignan como las entradas de la tabla.

```python
pd.DataFrame({ "Columna1":[1,2,3], "Columna2":[4,5,6]})
```

Una Serie no es mas que una **lista/secuencia de elementos**. \
Tambien se puede ver como la columna de un Dataframe. \
Podemos decir que un dataframe es la union de distintas Series.

```python
pd.Series([1, 2, 3, 4, 5])
```

El atributo _index_ especifica etiquetas para las filas que por defecto se asignan desde 0 en adelante

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

Podemos crear dataframe leyendo archivos externos, tales como excel, csv, sqlite, etc.

```python
pd.read_excel("Pokemon_quest_stones.xlsx")
```

## Indexar, seleccionar y manipular elementos

Seleccionar valores específicos de un DataFrame o Serie de pandas para trabajar es un paso implícito en casi cualquier operación de datos que ejecutará, por lo que una de las primeras cosas que debe aprender al trabajar con datos en Python es cómo seleccionar los puntos de datos relevantes para usted de manera rápida y efectiva.

### Selección nativa

De forma nativa cada columna del dataframe es un **atributo**

```python
df.columns
```

Podemos recuperar toda la columna accediendo a dicho atributo

```python
df.rating
```

Tambien podemos utilizar el **operador de indexacion []**

```python
df["rating"]
```

Este último tiene la ventaja de que podemos seleccionar datos en específico usando un segundo operador de indexación, siempre teniendo en cuenta que **primero va la columna y luego la fila**.

```python
df["name"][1]
```

Incluso hacer slicing con el inicio inclusivo y el final exclusivo

```python
df["name"][0:6]
```

### Método iloc
De forma alterma podemos usar el **paradigma de selección basada en indices** seleccionando FILAS de datos en base a su **posicion numérica** usando el método **iloc**.

```python
#Esto devuelve la primer fila
df.iloc[0]
```

```python
#Esto devuelve el elemento en la fila 1 columna 2 (atributo name)
df.iloc[0,1]
```

```python
#Podemos seguir haciendo slicing
df.iloc[0:6]
```

```python
df.iloc[0:6 , 1]
```

```python
#incluso seleccionar todas las filas en la penúltima columna
df.iloc[:,-2]
```

### Método loc
Por último podemos usar el **paradigma de selección basada en etiquetas** seleccionando FILAS de datos en base al **nombre de la columna** usando el método **loc**.

```python
df.loc[1 , "name"]
```

```python
df.loc[: , "rating"]
```

Si establecemos un rango de indices tanto el inicio como el final son inclusivos

```python
df.loc[1,"name":"rating"]
```

O podemos seleccionar solo algunas columnas

```python
df.loc[1 , ["name","rating"]]
```

Para ayudarnos a realizar mejor el paradigma de seleccion de etiquetas podemos cambiar los indices con el método set_index y almacenar el dataframe que devuelve en otro.

```python
new_df = df.set_index("name")
new_df.loc["Kimi no Na wa.","genre"]
```

### Selección condicional
Para realizar selecciones mas enriquecidas podemos aplicar condiciones.

```python
df.rating > 9
```

El resultado es una Serie de valores booleanos donde se asigna True aquellos que cumplen con la condición y False en caso contrario. \
Si asignamos esto al método loc obtenemos un dataframe solo con las filas que cumplen la condición.

```python
df.loc[df.rating > 9]
```

Podemos utlizar los operadores lógicos & y |

```python
df.loc[(df.rating<9) & (df.rating>8.5)]
```

```python
df.loc[(df.rating==8) | (df.rating==9)]
```

Podemos usar el metrodo isin para seleccionar datos donde los valores conciden con algún elemento de una lista de valores.

```python
df.loc[df.genre.isin(["Comedy","Action"])]
```

Por último, podemos modificar los datos como si de un diccionario se tratase.

```python
#Esto va a modificar todas las filas de la columna name
df["name"]="pepe"
df.head()
```

```python
#Con iloc es mas sencillo modificar entradas individiales por coordenadas
df.iloc[2,1]="Morita"
df.head()
```

```python
#Con loc es mas sencillo modificar entradas individuales por etiquetas, con la desventaja de trabajar con una copia del dataframe
copy = df.set_index("name")
copy.loc["Morita","genre"]="pepe le pu"
copy.head()
```

## Summary functions and maps
Los datos no siempre se seleccionan de la manera que los necesitamos, a veces tenemos que reformatearlos.

```python
#Primero volveremos a definir el dataframe
df = pd.read_csv("animes.csv")
df.head()
```

### Summary functions
Pandas nos proporciona distintas "summary functions" que pueden realizar este trabajo.

El método **describe** nos brinda un resumen de algunos atributos de la columna seleccionada.\
Este método es **type-aware** lo que significa que la salida depende del tipo de dato de la entrada.

```python
df.rating.describe()
```

```python
df.type.describe()
```

Puedes obtener solo algunas de estas propiedades (solo estadisticas) si las llamas por separado.

```python
df.rating.mean()
```

```python
df.type.unique()
```

```python
#Aqui puedes ver la frecuencia de las entradas unicas
df.type.value_counts()
```

### Método map
Map es un método que **designa una función sobre un conjunto de valores y los "mapea" en otro conjunto de valores**.

Las caracteristicas del método map son las siguientes:
- Se utiliza para crear nuevas representaciones de los datos a partir de los datos existentes.
- Se recomienda usar sobre las columnas del dataframe ya que si se hace sobre el dataframe completo la funcion modificara todas las columnas.
- Recibe una funcion que espera un único valor de la serie (columna del dataframe) y devuelve una versión tranformada de dicho valor.

```python
df.rating.map(lambda x : f"{x*10:.2f} points")
```

Apply es un método equivalente si queremos modificar toda la fila del dataframe aplicando un método distinto a cada columna.
